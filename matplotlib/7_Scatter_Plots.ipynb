## Lets plot Likes vs View count for few YouTube videos. 
The color of each point will be based on the Like/Dislike ratio of that video.

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('data/7_scatter_plots_data.csv')

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

# c: value for color intensity
# cmap: color map (black and white by default)
# edgecolor: edge around the points
# linewidth: width of edge
# alpha: adding transparency
plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

# display colorbar on the side to see color distribution
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

# plotting in logarithmic scale
plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
